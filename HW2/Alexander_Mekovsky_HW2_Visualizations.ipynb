```python
import pandas as pd
from matplotlib import pyplot as plt
from matplotlib.pyplot import figure
```

```python
data = pd.read_csv('scores.csv')
data
```

```python
acc = data['Accuracy']
f1 = data['F1']
f1 = f1 * 100
train = data['Train_Loss']
val = data['Val_Loss']
```

```python
figure(figsize = (16, 8))
plt.plot(acc)
plt.plot(f1)
plt.xlabel('Epochs')
plt.ylabel('Score')
plt.legend(['Accuracy','F1'])
plt.show()
```

```python
figure(figsize = (16, 8))
plt.plot(train)
plt.plot(val)
plt.xlabel('Epochs')
plt.ylabel('Score')
plt.legend(['Train_Loss','Val_Loss'])
plt.show()
```
